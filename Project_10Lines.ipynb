##### SELECT ROWS AND COLUMNS:

```python
'''select single columns'''
display(df['col1'])
'''Select multiple columns:'''
display(df[['col1','col2']])
'''Show first n rows'''
display(df.head(2))
'''Show last n rows'''
display(df.tail(2))
'''Select rows by index values'''
display(df.loc['A'])
display(df.loc[['A','B']])
'''Select rows by position'''
display(df.iloc[1])
display(df.iloc[1:])
```

##### DATA WRANGLING:

```python
'''filter by value'''
display(df[df['col1']>1])
'''sort by columns'''
display(df.sort_values(['col2','col2'],
               ascending=[False,True]))
'''Identify duplicate rows'''# hàng trùng lặp
display(df.duplicated())
'''Identify unique rows'''# Giá trị duy nhất
display(df['col1'].unique())
'''Swap rows and columns'''
df1=df.transpose()
df2=df.T
display(df2)
display(df1)
'''Drop a columns'''
df3=df.drop('col1',axis=1)
display(df3)
'''Clone a data frame'''
clone=df.copy()
print(clone)
'''Connect multiple data frames vertically'''
df4=df+5
display(df4)
pd.concat([df,df4])
```

```python
#Merge multiple data frames horizontally:
df3=pd.DataFrame([[1,7],[8,9]],
                 index=['B','D'],
                 columns=['col1','col3'])
print(df3)
#Only merge complete rows (INNER JOIN)
df4=df.merge(df3)
print(df4)
#Left columns stays complete (LEFT OUTER JOIN)
df_left=df.merge(df3,how='left')
display(df_left)
#Left columns stays complete (RIGHT OUTER JOIN)
df_right=df.merge(df3,how='right')
display(df_right)
#Preserve all values (OUTER JOIN)
df_outer=df.merge(df3,how='outer')
display(df_outer)
#Merge rows bt index
df.merge(df3,left_index=True,right_index=True)
#Fill NaN values:
df.fillna(0)
#Apply your own function:
def func(x):
  return 2**x
df_f=df.apply(func)
display(df_f)
```

##### Hierarchical indexing:

* Hierarchical indexing (hay còn được gọi là multi-indexing) là một cách sắp xếp dữ liệu trong pandas (một thư viện phân tích dữ liệu trong Python) dựa trên việc sử dụng nhiều cấp độ chỉ mục (index). Thông thường, chỉ mục trong pandas là một danh sách các nhãn (labels) cho các hàng (rows) hoặc cột (columns) của DataFrame hoặc Series. Khi sử dụng hierarchical indexing, mỗi mục trong chỉ mục có thể chứa nhiều nhãn (labels), dẫn đến việc tạo ra các bộ đếm (tuples) của nhãn để đại diện cho một mục.

* Hierarchical indexing cho phép ta đại diện cho dữ liệu dưới dạng các bảng dữ liệu (data tables) phức tạp hơn, bao gồm các cấp độ dữ liệu (levels of data) khác nhau và giúp cho việc truy xuất và xử lý dữ liệu trở nên dễ dàng hơn. Bằng cách sử dụng hierarchical indexing, ta có thể tạo ra các báo cáo tập trung vào các nhóm dữ liệu (grouped data reports) hoặc tính toán trên các bộ phận của dữ liệu một cách nhanh chóng và hiệu quả hơn.






```python
print(df.stack())#Create hierarchical index
print(df.unstack())# Dissolve hierarchical index
```

##### AGGREGATION:

* Aggregation là một khái niệm trong phân tích dữ liệu (data analysis) và thường được sử dụng để tính toán các thống kê tổng hợp (summary statistics) từ dữ liệu có sẵn. Cụ thể hơn, aggregation là quá trình tính toán các thống kê tổng hợp của một tập dữ liệu (dataset) như trung bình, độ lệch chuẩn, tổng, số lượng, giá trị lớn nhất, giá trị nhỏ nhất, v.v.

* Trong pandas, aggregation thường được sử dụng kết hợp với groupby, là một công cụ phân nhóm dữ liệu, để thực hiện các tính toán tổng hợp trên các nhóm dữ liệu riêng biệt. Ví dụ, ta có thể sử dụng aggregation để tính toán trung bình giá trị của từng nhóm sản phẩm trong một bảng dữ liệu về bán hàng.

* Aggregation cũng có thể được sử dụng để đối chiếu các giá trị dữ liệu với một mức độ tổng quát hơn, ví dụ như tính toán trung bình giá trị trong một khoảng thời gian (time period) hoặc tính toán tổng số sản phẩm được bán ra trong một khu vực địa lý cụ thể.

* Aggregation là một công cụ quan trọng trong phân tích dữ liệu, giúp chúng ta hiểu rõ hơn về tính chất của dữ liệu và tìm ra những mẫu chính xác trong dữ liệu để đưa ra các quyết định thông minh.

```python
g=df.groupby('col1')#Create group object
for i, group in g:#Iterate over groups
  print(i,group)
#Aggregate groups:
print(g.sum())
print(g.prod())
print(g.mean())
print(g.std())
print(g.describe())
```

```python
#Select columns from groups:
print(g['col2'].sum())
print(g[['col1','col2']].sum())
#Transform values:
import math
g.transform(math.log)
```
