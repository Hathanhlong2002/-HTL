### Từ điển -Dictionaries:

#### phương pháp từ điển
* Python có một tập hợp các phương thức tích hợp sẵn mà bạn có thể sử dụng trên từ điển.

* Method	Description
* clear()	Removes all the elements from the dictionary
* copy()	Returns a copy of the dictionary
* fromkeys()	Returns a dictionary with the specified keys and value
* get()	Returns the value of the specified key
* items()	Returns a list containing a tuple for each key value pair
* keys()	Returns a list containing the dictionary's keys
* pop()	Removes the element with the specified key
* popitem()	Removes the last inserted key-value pair
* setdefault()	Returns the value of the specified key. If the key does not exist: insert the key, * * with the specified value
* update()	Updates the dictionary with the specified key-value pairs
* values()	Returns a list of all the values in the dictionary

```python
"""Từ điển"""
adict={'a':'1',
      'b':'2',
      'c':'3',
      'y':192}
"""In từ điển và vị trí trong nó"""
print( adict)
print('vị trí key a là '+adict['a'])
print('lấy giá trị của key b: '+adict.get('b'))
"""Cách lấy danh sách khóa và giá trị của từ điển"""
print(adict.keys())
print(adict.values())
print(adict.items())# Danh sách các cặp key và values
"""Độ dài và kiểu dữ liệu của từ điển:"""
print(len(adict))
print(type(adict))
"""Thêm và cập nhật dữ liệu"""
adict['new']='onnew'
print(adict)
adict.update({'c':'8'})
print(adict)
"""Kiểm tra khóa có tồn tại trong từ điển"""
if "a" in adict:
      print('key a có tồn tại')
"""Xóa mục với tên khóa chỉ định trong từ điển:"""
adict.pop('new')
print(adict)
del adict['b']#Xóa từ điển key b, có thể dùng xóa cả từ điển
print(adict)
adict.popitem()#Xóa vị trí cuối cùng của từ điển
print(adict)
"""Làm trống từ điển"""
adict.clear()
print(adict)
""""Vòng lặp trong từ điển:"""
#Khai báo lại từ điển
adict={'a':'1',
      'b':'2',
      'c':'3',
      'y':192}
print('key: ')
for x in adict:
      print(x,end=' - ')
print(" ")
for x in adict.keys():
      print(x,end=' - ')
print('\nvalues:')
for i in adict.values():
      print(i,end='*  ')
print('')
for a in adict:
      print(adict[a],end='+  ')
"""Lặp qua khóa và giá trị của từ điển:"""
print('\nvalues and key:')
for x, y in adict.items():
  print(x, y,end='\t')
```

#### KẾ THỪA- INHERITANCE:

* Lớp cha: là lớp được kế thừa từ đó,là lớp cơ sở **Based class**
* Lớp con: là lớp kế thừa từ lớp khác, là lớp dẫn xuất **Derived class**
* **SUPER()** sẽ làm cho lớp con kế thừa tất cả các phương thức và thuộc tính từ cha của nó

```python
'''Tạo Lớp cha'''
class Person:
  def __init__(self,fname,lname,age):
    self.fname=fname
    self.lname=lname
    self.age=age
  def print_name(self):
    print(self.fname," ",self.lname,'age: ',self.age)

x=Person('alice','Boe',34)
x.print_name()
```

```python
'''Tạo lớp con'''
class Student(Person):
  def __init__(self,fname,lname,age,height):
    self.fname=fname
    self.lname=lname
    self.age=age
    self.height=height


y=Student('thane','roke',14,122)
y.print_name()
```

```python
'''Giữ tính kế thừa chức năng của lớp cha'''
class Student(Person):
  def __init__(self,fname,lname,age):
    Person.__init__(self,fname,lname,age)
```

```python
'''Dùng hàm super() giữ kế thừa tất cả phương thức của lớp cha 
và thêm thuộc tính cho lớp Student'''
class Student(Person):
  def __init__(self,fname,lname,age,year):
    super().__init__(fname,lname,age)
    self.gradeclass=year
  def welcome(self):
    print('Welcome',self.fname,self.lname,'have ',self.age,'born: ',self.gradeclass)
x=Student('name','my name',20,2002)
x.print_name()
x.welcome()
```

```python
'''Class'''
class MyNumber:
  def __iter__(self):
    self.a=1
    return self
  def __next__(self):
    if self.a<=20:
      x=self.a
      self.a+=1
      return x
    else:
      raise StopIteration
    
myNum=MyNumber()
myiter=iter(myNum)
for i in myiter:
  print(next(myiter),end=' ')
```

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 20:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration

myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
  print(x,end=' ')
```

### Date- Ngày giờ

* Sử dụng module: **datetime**
> **datetime.**
  - 1.datetime():
  - 2.datetime.now():
  - 3.strftime():
>> - 4.%a: Tên ngày trong tuần, phiên bản ngắn gọn. Ví dụ: Wed.
  - 5.%A: Tên ngày trong tuần, phiên bản đầy đủ. Ví dụ: Wednesday.
  - 6.%w: Số thứ tự của ngày trong tuần, bắt đầu từ 0 cho Chủ nhật. Ví dụ: 3.
%d: Ngày trong tháng, dạng số, từ 01 đến 31.
  - 7.%b: Tên tháng, phiên bản ngắn gọn. Ví dụ: Dec.
  - 8.%B: Tên tháng, phiên bản đầy đủ. Ví dụ: December.
  - 9.%m: Tháng, dạng số, từ 01 đến 12.
  - 10.%y: Năm, phiên bản ngắn gọn, không bao gồm thế kỷ. Ví dụ: 18.
  - 11.%Y: Năm, phiên bản đầy đủ. Ví dụ: 2018.
  - 12.%H: Giờ, dạng số, từ 00 đến 23.
  - 13.%I: Giờ, dạng số, từ 00 đến 12.
  - 14.%p: Thời điểm trong ngày, AM hoặc PM.
  - 15.%M: Phút, dạng số, từ 00 đến 59.
  - 16.%S: Giây, dạng số, từ 00 đến 59.
  - 17.%f: Micro giây, dạng số, từ 000000 đến 999999.
  - 18.%z: Khoảng cách thời gian đến giờ chuẩn phối hợp thế giới (UTC).

### Mảng sao chép và xem mảng: COPY và VIEW()

* Sự khác biệt chính giữa bản sao và dạng xem của một mảng là bản sao là một mảng mới và dạng xem chỉ là dạng xem của mảng ban đầu.

#### COPY():

```python
import numpy as np 
'''Mảng COPY sẽ không ảnh hưởng đến mảng gốc'''
arr=np.array([1,2,3,4,5,6,5,6,4])
b=arr.copy()
arr[0]=9

print(arr)
print(b)
```

#### VIEW():

```python
import numpy as np
'''Mảng VIEW() sẽ bị thay đổi khi mảng gốc thay đổi và ngược lại'''
arr=np.array([2,4,0,3,5,6,2])
b=arr.view()
arr[0]=8

print(arr)
print(b)

b[3]=33
print(arr)
print(b)
```

### Mảng LẶP LẠI:Array Iterating

```python
import numpy as np

arr=np.array([1,2,3])
for x in arr:
  print(x,end=' ')
print('\n')

arr2D=np.array([[1,2,3,4],[1,4,5,6]])
for i in arr2D:
  print(i)

for i in arr2D:
  for a in i:
    print(a,end=' - ')
```

### MẢNG THAM GIA: JOINING ARRAY:

* Trong **SQL** nối các bảng dựa trên một khóa, còn **NUMPY** thì nối mảng theo [ **trục** ]
* Sử dụng hàm **CONCATENATE()** để nối một chuỗi các mảng

```python
import numpy as np
'''2 Mảng'''
arr1=np.array([1,2,3])
arr2=np.array([3,4,5])

arr_concat2=np.concatenate((arr1,arr2))
print(arr_concat2)
"""2D với trục bằng 1"""
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[5,6],[7,8]])
arr_concat2D=np.concatenate((arr1,arr2),axis=0)# axis chỉ có giá trị 0 và 1
print('\n')
print(arr_concat2D)
```

#### Tham gia mảng sử dụng hàm **ngăn xếp** ***( STACK FUNCTIONS)***

* Sử dụng xếp chồng các mảng theo một trục mới
VD: Mảng 1D thì đặt chúng chồng lên nhau:
Hàm concatenate() nối chúng vào chung một mảng, còn hàm stack() xếp chồng các mảng lên nhau.

```python
import numpy as np

arr1=np.array([1,2,3,4])
arr2=np.array([2,3,5,6])
arr_stack1=np.stack((arr1,arr2),axis=1)
print(arr_stack1)

arr_stack2=np.stack((arr1,arr2))
print(arr_stack2)
```

### TÌM KIẾM MẢNG: NumPy Searching Arrays

* Để tìm kiếm một mảng sử dụng hàm **WHERE()**
* Hàm **SEARCHSORTED()** là thực hiện tìm kiếm nhị phân trong mảng và trả về chỉ mục nơi giá trị đã chỉ định sẽ được chèn vào để duy trì thứ tự tìm kiếm
> **side='right'** tìm kiếm từ bên phải

```python
import numpy as np

arr=np.array([1,2,3,4,5,5,6,7,8,8,1,2,12,21,14,1,6])
'''Tìm kiếm các vị trí mà có giá trị bằng 1'''
x=np.where(arr==1)
print(x)
'''Tìm các vị trí mà là số chẵn'''
y=np.where(arr%2==0)
print(y)
'''Tìm vị trí mà giá trị 5 được chèn vào, tìm kiếm từ trái qua'''
q=np.searchsorted(arr,5)
print(q)
'''tìm kiếm vị trí mà giá trị 14 được chèn vào bắt đầu từ bên phải'''
p=np.searchsorted(arr,14,side='right')
print(p)
'''tìm kiếm một mảng các giá trị '''
a=np.searchsorted(arr,[1,3,5,6,9])
print(a)
```

### Simple Arithmetic: Số học đơn giản:

* Hàm **add()**tính tổng nội dung của hai mảng và trả về kết quả trong một mảng mới.
* Hàm **subtract()**trừ các giá trị từ một mảng với các giá trị từ một mảng khác và trả về kết quả trong một mảng mới.
* Hàm **multiply()**nhân các giá trị từ một mảng với các giá trị từ một mảng khác và trả về kết quả trong một mảng mới.
* Hàm **divide()**chia các giá trị từ một mảng với các giá trị từ một mảng khác và trả về kết quả trong một mảng mới.

* Hàm **divmod()**trả về cả thương và **mod**. Giá trị trả về là hai mảng, mảng đầu tiên chứa thương và mảng thứ hai chứa mod.
* Cả hàm **absolute()** và **abs()** hàm đều thực hiện cùng một yếu tố hoạt động tuyệt đối nhưng chúng ta nên sử dụng **absolute()** để tránh nhầm lẫn với hàm sẵn có của **pythonmath.abs()**

```python
import numpy as np

arr1 = np.array([10, 11, 12, 13, 14, 15])
arr2 = np.array([20, 21, 22, 23, 24, 25])

add_arr=np.add(arr1,arr2)
print(add_arr)
subtract_arr=np.subtract(arr1,arr2)
print(subtract_arr)
mul_arr=np.multiply(arr1,arr2)
print(mul_arr)
divide_arr=np.divide(arr1,arr2)
print(divide_arr)

pow_arr=np.power(arr1,arr2)
print(pow_arr)
mod_arr = np.mod(arr1, arr2)
print(mod_arr)
remain_arr = np.remainder(arr1, arr2)
print(remain_arr)
div_arr = np.divmod(arr2, arr1)
print(div_arr)
```

### Rounding Decimals: Làm trònv số thập phân:

> * Xóa các số thập phân và trả về số float gần nhất với 0. Sử dụng các chức năng **trunc()** và **fix()**
> * Hàm **around()** tăng số trước chữ số hoặc số thập phân lên 1 nếu >=5 khác thì không làm gì cả.
> * Hàm **floor()** làm tròn số thập phân thành số nguyên thấp hơn gần nhất.
> * Hàm **ceil()** làm tròn số thập phân thành số nguyên trên gần nhất.

```python
import numpy as np

arr1 = np.trunc([-3.1666, 3.6667]);print(arr1)
arr2 = np.fix([-3.1666, 3.6667]);print(arr2)
arr3 = np.around(3.1666, 2);print(arr3)
arr4 = np.floor([-3.1666, 3.6667]);print(arr4)
arr5 = np.ceil([-3.1666, 3.6667]);print(arr5)
```

### DATA CLEANING: Làm sạch dữ liệu:

* Làm sạch dữ liệu có nghĩa là sửa dữ liệu xấu trong tập dữ liệu:
* Dữ liệu xấu có thể là:
    * Ô trống **EMPTY CELLS**
    * Dữ liệu sai định dạng 
      > + **DATA IN WRONG FORMAT**
    * Dữ liệu sai **WRONG DATA** 
    * Trùng lặp **DUPLICATES**
* Xóa hàng:
    * Cách xử lý các ô trống là loại bỏ các hàng chứa ô trống đó
    * Vì tập dữ liệu sẽ rất lớn nên việc xóa vài hàng sẽ không ảnh hướng để kết quả
    * Dùng **DROPNA()** để thực hiện và phương thức này sẽ trả về một **Dataframe** mới và không thay đổi bản gốc
        * Muốn thay đổi cả dataframe ban đầu hãy sử dụng **INPLACE=TRUE**
* Thay thế các giá trị rỗng:
    * Một cách khác để xử lý các ô trống là thay vào đó bằng cách chèn một giá trị mới
    * Dùng **FILLNA()** cho phép thay thế các ô trống bằng một giá trị
    > * Có thể thay thế các giá trị **NULL** ở các cột chỉ định
* Thay thế bằng **MEAN, MEDIAN,MODE**:
    * Dùng **MEAN(), MEDIAN(),MODE():** để tính toán các giá trị tương ứng cho một cột chỉ định
* Làm sạch dữ liệu sai định dạng: WRONG FORMAT:
    * Có 2 cách để khắc phục:
        * Loại bỏ các hàng 
        * Chuyển các ô trong các cột về cùng định dạng
    * Phương thức **TO_DATETIME()** định dạng ngày cho các cột có thời gian
* Sửa dữ liệu sai: WRONG DATA:
    * Dữ liệu sai không nhất thiết phải là dữ liệu rỗng hoặc là có định dạng sai, nó chỉ có thể sai:)))
    * Các cách khắc phục:

    * Để loại bỏ trùng lặp, sử dụng **DROP_DUPLICATES()**

### Optimizers:Tối ưu hóa scipy

* Trình tối ưu hóa trong scipy:
  * Là một tập hợp các quy trình được xác định trong scipy để tìm giá trị nhỏ nhất của một hàm hoặc nghiệm của một phương trình
  * Về cơ bản, tất cả các thuật toán trong Machine Learning không gì khác hơn là một phương trình phức tạp cần được giảm thiểu với sự trợ giúp của dữ liệu đã cho.
* **NUMPY** có khả năng tìm nghiệm cho đa thức và **phương trình tuyến tính**, nhưng nó không thể tìm nghiệm cho phương trình **phi tuyến tính**
  * Để thực hiện được, dùng **OPTIMZE.ROOT** của SCIPY
    * Gồm 2 đối số:
      * **FUN**: một chức năng đại diện cho một phương trình
      * **X0** dự đoán ban đầu cho **gốc**
    * Hàm trả về một đối tượng có thông tin liên quan đến giải pháp
* Giảm thiểu chức năng:
  * Một hàm đại diện cho một đường cong, các đường cong có điểm cao và điểm thấp
    * Điểm cao là điểm cực đại
    * Điểm thấp là điểm cực tiểu
  * Điểm cao nhất trong toàn bộ đường cong được gọi là **cực đại toàn cầu** , trong khi các điểm còn lại được gọi là **cực đại cục bộ** 
  * Điểm thấp nhất trong toàn bộ đường cong được gọi là **cực tiểu toàn cầu** , trong khi các điểm còn lại được gọi là **cực tiểu cục bộ** .
* Tìm cực tiểu:
  * Dùng **SCIPY.OPTIMIZE.MINIMIZE()** để tối thiểu hóa chức năng
  * Hàm **MINIMIZE()** gồm các đối số:
    * **FUN** chức năng đại diện cho phương trình
    * **x0** dự đoán ban đầu cho gốc
    * **METHOD** tên method sử dụng. Gồm các giá trị pháp lý:
      *  **'CG'**
      *  **'BFGS'**
      *  **'Newton-CG'**
      *  **'L-BFGS-B'**
      *  **'TNC'**
      *  **'COBYLA'**
      *  **'SLSQP**
    * **CALLBACK**:Chức năng được gọi lại sau mỗi lần lặp tối ưu hóa
    * **OPTIONS**: một từ điển xác định các thông số bổ sung:
      * **{"disp": boolean - print detailed description**
     * **"gtol": number - the tolerance of the error }**

```python
from scipy.optimize import root
from math import cos

def eqn(x):
  return x+cos(x)
'''Tìm nghiệm của phương trình'''
myroot=root(eqn,0)
print(myroot.x)
'''in tất cả các thông tin về giải pháp'''
print(myroot)
```

```python
from scipy.optimize import minimize

def eqn(x):
  return x**2+x+2

mymin=minimize(eqn,0,method='BFGS')

print(mymin)
```

### SPARSE: Dữ liệu thưa thớt:

* Dữ liệu thưa thớt là dữ liệu có hầu hết các phần tử không được sử dụng ( Các phần tử không mang bất kỳ thông tin nào) nghĩa là hầu hết các giá trị đều bằng 0
*  Mảng dày đặc ngược lại, hầu hết các giá trị khác 0
* Dùng **SCIPY.SPARSE** cung cấp các chức năng để xử lý dữ liệu thưa thớt
* Chủ yếu có hai loại ma trận thưa thớt mà chúng tôi sử dụng:
  * **CSC** - Cột thưa được nén. Để tính toán số học hiệu quả, cắt cột nhanh.
  * **CSR** - Hàng thưa được nén. Để cắt hàng nhanh, các sản phẩm vectơ ma trận nhanh hơn
* Đếm các số 0 bằng **COUNT_NONZERO()** 
* Loại bỏ các mục nhập bằng 0 khỏi ma trận **ELIMINATE_ZEROS()**
* Loại bỏ các mục trùng lặp bằng **SUM_DUPLICATES()** 
* Chuyển từ **CSR** sang **CSC** bằng **TOCSC()**

```python
import numpy as np
from scipy.sparse import csr_matrix
'''scipy.sparse.csr_matrix()'''
arr = np.array([0, 0, 0, 0, 0, 1, 1, 0, 2])
print(csr_matrix(arr))

arr1 = np.array([[0, 0, 0], [0, 0, 1], [1, 0, 2]])
print(csr_matrix(arr1).data)

print(csr_matrix(arr).count_nonzero())
print(csr_matrix(arr1).count_nonzero())

mat = csr_matrix(arr)
mat.eliminate_zeros()
print(mat)

mat.sum_duplicates()
print(mat)

newarr = csr_matrix(arr).tocsc()
print(newarr)
```

#### Dijkstra:

* Sử dụng **dijkstra** phương pháp tìm đường đi ngắn nhất trong một đồ thị từ phần tử này đến phần tử khác.
* Nó có các đối số sau:
  * **return_predecessors**: boolean (Đúng để trả về toàn bộ đường dẫn truyền tải nếu không là Sai).
  * **indices**: chỉ mục của phần tử để chỉ trả về tất cả các đường dẫn từ phần tử đó.
  * **limit**: trọng lượng tối đa của đường dẫn.

```python
import numpy as np
from scipy.sparse.csgraph import dijkstra
from scipy.sparse import csr_matrix

arr = np.array([
  [0, 1, 2],
  [1, 0, 0],
  [2, 0, 0]
])

newarr = csr_matrix(arr)
print(dijkstra(newarr, return_predecessors=True, indices=0))
```

#### Floyd Warshall:

* Dùng **floyd_warshall()** để tìm đường đi ngắn nhất giữa tất cả các cặp phần tử

```python
import numpy as np
from scipy.sparse.csgraph import floyd_warshall
from scipy.sparse import csr_matrix

arr = np.array([
  [0, 1, 2],
  [1, 0, 0],
  [2, 0, 0]
])

newarr = csr_matrix(arr)

print(floyd_warshall(newarr, return_predecessors=True))
```

#### Bellman Ford:

* Phương thức này **bellman_ford()**cũng có thể tìm đường đi ngắn nhất giữa tất cả các cặp phần tử, nhưng phương thức này cũng có thể xử lý các trọng số âm.

```python
import numpy as np
from scipy.sparse.csgraph import bellman_ford
from scipy.sparse import csr_matrix

arr = np.array([
  [0, -1, 2],
  [1, 0, 0],
  [2, 0, 0]
])

newarr = csr_matrix(arr)

print(bellman_ford(newarr, return_predecessors=True, indices=0))
```

#### Depth First Order:Độ sâu thứ tự đầu tiên:

* Phương thức này **depth_first_order()**trả về một lần duyệt theo chiều sâu đầu tiên từ một nút.
* Hàm này nhận các đối số sau:
  * đồ thị **graph**
  * phần tử bắt đầu để đi qua biểu đồ từ đó.

```python
import numpy as np
from scipy.sparse.csgraph import depth_first_order
from scipy.sparse import csr_matrix

arr = np.array([
  [0, 1, 0, 1],
  [1, 1, 1, 1],
  [2, 1, 1, 0],
  [0, 1, 0, 1]
])

newarr = csr_matrix(arr)
print(depth_first_order(newarr, 1))
```

#### Breadth First Order: Chiều rộng thứ tự đầu tiên:

* Hàm này nhận các đối số sau:
  * đồ thị.
  * phần tử bắt đầu để đi qua biểu đồ từ đó.

```python
import numpy as np
from scipy.sparse.csgraph import breadth_first_order
from scipy.sparse import csr_matrix

arr = np.array([
  [0, 1, 0, 1],
  [1, 1, 1, 1],
  [2, 1, 1, 0],
  [0, 1, 0, 1]
])

newarr = csr_matrix(arr)

print(breadth_first_order(newarr, 1))
```

  * Một tam giác với các điểm có nghĩa là tạo ra các tam giác bao gồm bề mặt trong đó tất cả các điểm đã cho nằm trên ít nhất một đỉnh của tam giác bất kì trên bề mặt

```python
import numpy as np
from scipy.spatial import Delaunay
import matplotlib.pyplot as plt

points = np.array([
  [2, 4],
  [3, 4],
  [3, 0],
  [2, 2],
  [4, 1]
])

simplices = Delaunay(points).simplices

plt.triplot(points[:, 0], points[:, 1], simplices)
plt.scatter(points[:, 0], points[:, 1], color='r')
''' Thuộc tính simplices tạo ra một tổng quát của ký hiệu tam giác'''
plt.show()
```

```python
import numpy as np
from scipy.spatial import ConvexHull
import matplotlib.pyplot as plt

points = np.array([
  [2, 4],
  [3, 4],
  [3, 0],
  [2, 2],
  [4, 1],
  [1, 2],
  [5, 0],
  [3, 1],
  [1, 2],
  [0, 2]
])

hull = ConvexHull(points)
hull_points = hull.simplices

plt.scatter(points[:,0], points[:,1])
for simplex in hull_points:
  plt.plot(points[simplex,0], points[simplex,1], 'k-')

plt.show()
```

### SCiPY INTERPOLATION: Nội suy

* Nội suy là phương pháp để tạo điểm giữa các điểm đã cho:
* Nội suy có nhiều cách sử dụng:
  * thường được xử lý dữ liệu bị thiếu trong tập dữ liệu, phép nội suy dùng để thay thế các giá trị đó
  * **=> QUY NẠP**
* Dùng **MODULE SCIPY.INTERPOLATE** có các chức năng để xử lý phép nội suy:

#### NỘI SUY 1D:

* Hàm **INTERP1D()** được sử dụng để nội suy phân phối 1 biến
  * Nhận x,y và trả về hàm có thể gọi được trả về giá trị x,y **mới**

```python
from scipy.interpolate import interp1d
import numpy as np

xs=np.arange(10)
ys=2*xs+1

interp_func=interp1d(xs,ys)
newarr=interp_func(np.arange(2.1,3,0.1))
print(newarr)
```

#### Nội suy SPLINE:

* Trong nội suy 1D, các điểm được khớp với một đường cong duy nhất
* Phép nội suy Spline các điểm được khớp với một hàm từng phần được xác định bằng đa thức là các đường spline
* Hàm **UNIVARIATESPLINE()** nhận xs và ys tạo ra một hàm mới

```python
from scipy.interpolate import UnivariateSpline
import numpy as np

xs=np.arange(10)
ys=xs**2+np.sin(xs)+1

interp_func=UnivariateSpline(xs,ys)
newarr=interp_func(np.arange(2.1,4,0.1))
print(newarr)
```

#### Interpolation with Radial Basis FUNCTION: Nội suy với chức năng cơ sở radius

* Được xác định tương ứng với một tham chiếu cố Định
* Hàm **RBF()** nhận xs và ys làm đối số tạo hàm mới

```python
from scipy.interpolate import Rbf
import numpy as np
xs = np.arange(10)
ys = xs**2 + np.sin(xs) + 1

interp_func = Rbf(xs, ys)

newarr = interp_func(np.arange(2.1, 3, 0.1))

print(newarr)
```

### SCIPY satistical significance Tests: Kiểm tra ý nghĩa thống kê:

* Kiểm tra ý nghĩa thống kê có nghĩa là kết quả được tạo ra đằng sau nó, không được tạo ra một cách ngẫu nhiên hoặc tình cờ
* Scipy cung cấp một module **SCIPY.STATS**: có chức năng thực hiện các bài kiểm tra ý nghĩa thống kê

*   Giả thuyết trong thống kê Giả thuyết là một giả định về một tham số trong dân số.  giả thuyết không Nó giả định rằng quan sát không có ý nghĩa thống kê.
* giả thuyết thay thế
  * Nó giả định rằng các quan sát là do một số lý do.
  * Nó thay thế cho Giả thuyết Null.
* Bài kiểm tra một đuôi
  * Khi giả thuyết của chúng tôi chỉ kiểm tra một mặt của giá trị, nó được gọi là "kiểm tra một đuôi".
* Bài kiểm tra hai đuôi
  * Khi giả thuyết của chúng tôi đang thử nghiệm cho cả hai mặt của các giá trị.
* Giá trị alpha: là mức ý nghĩa
* Giá trị P: 
  * Cho biết mức độ gần với mức cực đoan của dữ liệu thực tế
  * Giá trị P và alpha được so sánh để thiết lập ý nghĩa thống kê
* Nếu giá trị p <= alpha, chúng tôi bác bỏ giả thuyết khống và nói rằng dữ liệu có ý nghĩa thống kê. mặt khác, chúng tôi chấp nhận giả thuyết khống.
* Giá trị T:
  * Thử nghiệm T được sử dụng để xác định xem có sự khác biệt đáng kể giữa giá trị trung bình của hai biến hay không và cho chúng tôi biết liệu chúng có thuộc cùng một phân phối hay không.
  * Đó là một bài kiểm tra hai đuôi.
  * Hàm **ttest_ind()**lấy hai mẫu có cùng kích thước và tạo ra một bộ thống kê t và giá trị p.






```python
import numpy as np
from scipy.stats import ttest_ind

v1 = np.random.normal(size=100)
v2 = np.random.normal(size=100)

res = ttest_ind(v1, v2)

print(res)
'''in ra giá trị P'''
res = ttest_ind(v1, v2).pvalue
print(res)
```

### Linear Regression: Thuật toán hồi quy tuyến tính:

* **Regression** thuật ngữ hồi quy được sử dụng khi cố gắng tìm kiếm mối quan hệ giữa các biến
* Dùng để dự đoán các sự kiện trong tương lai
* **Linear Regression** sử dụng mối quan hệ giữa các điểm dữ liệu để vẽ một đường thẳng qua tất cả các điểm đó
* Dùng để dự đoán các giá trị trong tương lai

* Hàm **map()** trong Python được sử dụng để áp dụng một hàm cho mỗi phần tử trong một iterable (như list, tuple, dictionary, set, ...) và trả về một iterator chứa các kết quả tương ứng.

```python
import matplotlib.pyplot as plt
from scipy import stats
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]# Đại diện cho số tuổi
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]# đại diện cho tốc độ

slope,intercept,r,p,std_err=stats.linregress(x,y)
#Hàm stats.linregress() được sử dụng để tính toán hệ số góc (slope), 
#hệ số chặn (intercept), hệ số tương quan (r), sai số tiêu chuẩn (std_err) 
#và giá trị p cho một mô hình hồi quy tuyến tính giữa x và y.

def myfunc(x):
  return slope*x+intercept
#Hàm myfunc() được định nghĩa để tính toán giá trị dự đoán y dựa trên giá trị x
# và hệ số góc và chặn được tính toán ở bước trước đó.
mymodel=list(map(myfunc,x))
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```

* Mối quan hệ R:
  * Điều quan trọng là phải biết mối quan hệ giữa các giá trị của trục x và y là như thế nào, nếu không có mối quan hệ nào thì hồi quy tuyến tính không thể được sử dụng để dự đoán
  * Hệ số tương quan gọi là **R**
  * Giá trị **R** nằm trong khoảng -1 đến 1, trong đó 0 nghĩa là không có mối quan hệ nào và 1 và -1 có nghĩa là 100% có mối liên hệ

```python
'''Dữ liệu có phù hợp trong quy hoạch tuyến tính'''
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope,intercept,r,p,std_err=stats.linregress(x,y)
print(slope)
print(r)
print(intercept)
```

### Polynomial Regression: hồi quy đa thức:

* Nếu dữ liệu không phù hợp với hồi quy tuyến tính khi đồ thị của nó là một đường thẳng qua tất cả điểm dữ liệu, thì nó có thể sẽ lý tưởng cho hồi quy đa thức:
* Cũng giống như hồi quy tuyến tính nó sử dụng các mối quan hệ giữa các biến x và y để tìm ra cách tốt nhất để vẽ một đường thẳng qua các điểm dữ liệu

```python
import numpy as np
import matplotlib.pyplot as plt

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]# số giờ trong ngày
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]# tốc độ

mymodel = np.poly1d(np.polyfit(x, y, 3))# Fit đa thức bậc 3 vào dữ liệu x,y

myline = np.linspace(1, 22, 100)# tạo 100 giá trị trong khoảng 1-22

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))# tính giá trị y tương ứng với mỗi x trên đường cong
plt.show()
```

* **R-SQUARE** là hệ số xác định : coefficient of determination là một đại lượng thống kê được sử dụng để đánh giá độ chính xác của một mô hình hồi quy.
* Nó được xác định bằng cách so sánh giá trị thực tế và giá trị được dự đoán bởi mô hình, với độ lệch trung bình của dữ liệu thực tế
* R-square có giá trị từ 0-1 và giá trị càng gần 1 thì mô hình càng tốt, nghĩa là mô hình hoàn toàn chính xác và không có sai số nào. nếu R-square gần bằng 0 nghĩa là mô hình không có khả năng dự đoán tốt hơn so với việc chọn giá trị trung bình cho tất cả các mẫu
* **NOTE:**: mô hình có R-square cao hơn thì là mô hình tốt hơn. Tuy nhiên , R-square cũng có một số hạn chế và không phải đại lượng duy nhất để đánh giá một mô hình hồi quy vì nó không cho biết độ lớn của sai số dự đoán , không thể xác định được tình trạng overfitting, và không phù hợp khi số lượng biến độc lập lớn .




```python
import numpy as np
from sklearn.metrics import r2_score

x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

mymodel=np.poly1d(np.polyfit(x,y,3))
print(r2_score(y,mymodel(x)))
```

### SCALE: Chuẩn hóa:

* Việc chuẩn hóa dữ liệu là một bước quan trọng để đảm bảo rằng các thuộc tính của dữ liệu có cùng thang đo, tránh hiện tượng một thuộc tính có giá trị quá lớn so với các thuộc tính khác và ảnh hưởng đến quá trình huấn luyện mô hình.
* Có nhiều phương pháp chuẩn hóa dữ liệu:
  * Standard Scaling: Chuẩn hóa dữ liệu bằng cách trừ giá trị trung bình của toàn bộ dữ liệu và chi cho độ lệch chuẩn. Phương pháp này đưa dữ liệu về dạng phân phối có giá trị bằng 0 và độ lệch chuẩn bằng 1
  * Min-Max Scaling: Chuẩn hóa dữ liệu bằng cách đưa các giá trị về khoảng cụ thể. Đưa dữ liệu về khoảng giá trị mong muốn và đồng nhất giá trị của các thuộc tính.

#### SCALE FEAURES: chuẩn hóa đặc trưng

* Việc so sánh giữa hai dữ liệu chênh lệch nhau lớn có thể rất khó , vì vậy cần đứa dữ liệu về giá trị có thể so sánh được, từ đó có thể thấy được mức độ khác biệt giữa hai dữ liệu đó.
* Có nhiều phương pháp đưa dữ liệu về một phạm vi so sánh được:
  * Sử dụng phương pháp chuẩn hóa:
    *      **z=(x-u)/s**
      * z: là giá trị mới
      * x: là giá trị ban đầu
      * u: là trung bình
      * s là độ lệch chuẩn
* Sử dụng module sklearn trong python có phương thức StandardScaler() để chuẩn hóa một tập dữ liệu.

```python
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

scale=StandardScaler()

df=pd.read_csv('data.csv')

X=df[['Weight','Volume']]
y=df['CO2']

scaledX=scale.fit_transform(X)
print(scaledX)

model=linear_model.LinearRegression()
model.fit(scaledX,y)

scaled=scale.transform([[23000,1.3]])

predictCo2=model.predict([scaled[0]])
print(predictCo2)
```
